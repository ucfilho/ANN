```python
stones2find = cv2.imread('20221103_143816quarter.jpg',0)
```

```python
plt.figure(figsize=(15, 15))
plt.imshow(cv2.cvtColor(cv2.imread('20221103_143816quarter.jpg'), cv2.COLOR_BGR2RGB))
```

```python
nrow, ncol = stones2find.shape
nrow, ncol
```

```python
size = 50
horiz = int(nrow/size)
vert = int(ncol/size)
segmentation = []
for i in range(horiz):
  h = i*size
  for j in range(vert):
    w = j*size
    img=np.zeros((size,size)).astype(np.int64)
    for ii in range(size):
      for jj in range(size):
        img[ii,jj] = stones2find[ii+h,jj+w]
    
    segmentation.append(img)



#plt.figure(figsize=(5,5))
#plt.imshow(img)
```
