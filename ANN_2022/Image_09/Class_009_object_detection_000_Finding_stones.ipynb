```python
stones2find = cv2.imread('20221103_143816quarter.jpg',0)
```

```python
plt.figure(figsize=(20, 20))
plt.imshow(cv2.cvtColor(cv2.imread('20221103_143816quarter.jpg'), cv2.COLOR_BGR2RGB))
```

```python
nrow, ncol = stones2find.shape
nrow, ncol
```

```python
width = int(ncol/20)
img=np.zeros((width,width)).astype(np.int64)
for i in range(width):
  for j in range(width):
       img[i,j] = stones2find[i,j]
plt.figure(figsize=(5,5))
plt.imshow(img)
```

```python
width = int(ncol/10)
w = 800
h = 2000
img=np.zeros((width,width)).astype(np.uint8)
for i in range(width):
  for j in range(width):
       img[i,j] = stones2find[i+h,j+w]
plt.figure(figsize=(5, 5))
plt.imshow(img)
```
