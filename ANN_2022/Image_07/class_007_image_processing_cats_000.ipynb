```python
img_name=[]
with zipfile.ZipFile(FILE, "r") as f:
    for name in f.namelist():
        img_name.append(name)
```

```python
#from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
X= []  
names =[]
Img_Size = 224
ref = 'jpg'
n = len(img_name)
i = 0
for i in range(n):
  name = img_name[i]
  if(ref in name):
    names.append(name)
    img =cv2.imread(name)
    #img = preprocess_input(img)
    X.append(cv2.resize(img, (Img_Size,Img_Size)))
```

```python
plt.figure(figsize=(15, 15))
n = len(names)
male, female = [],[]
for i in range(n):
  if('f' in names[i]):
    female.append(i)
  else:
    male.append(i)
lista = []
for i in range(len(female)):
  lista.append(female[i])
  lista.append(male[i])

for i in range(16): 
  ax = plt.subplot(4, 4, i + 1)
  k = lista[i]
  RGB_img = cv2.cvtColor(X[k], cv2.COLOR_BGR2RGB)
  resized = cv2.resize(RGB_img, (Img_Size,Img_Size))
  plt.imshow(resized)
  plt.title(names[k])
  plt.axis("off")
```

```python
Y = []
for yi in names:
  if('f' in yi):
    Y.append(1)
  else:
    Y.append(0)
```
