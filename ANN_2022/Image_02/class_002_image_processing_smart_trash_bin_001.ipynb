```python
FILE='trash_nov_22_2018.zip'
```

```python
archive = zipfile.ZipFile('trash_nov_22_2018.zip', 'r')
archive.extractall()
```

```python
img_name=[]
with zipfile.ZipFile(FILE, "r") as f:
    for name in f.namelist():
        img_name.append(name)
```

```python
Img_Size = 80
ref = 'jpg'
notref = 'met'
```

```python
X= []  
Y =[]
n = len(img_name)
i = 0
for i in range(n):
  name = img_name[i]
  if(ref in name):
    Y.append(name)
    X.append(cv2.imread(name,0))
m = len(Y)
print(n,m)
```

```python
X= []  
Y =[]
n = len(img_name)
i = 0
for i in range(n):
  name = img_name[i]
  if(ref == name[-3:]):
    Y.append(name)
    X.append(cv2.imread(name,0))
m = len(Y)
print(n,m)
```

```python
plt.figure(figsize=(10, 10))
X_test = X
Y_test = Y
for i in range(16): 
  ax = plt.subplot(4, 4, i + 1)
  RGB_img = cv2.cvtColor(X_test[i], cv2.COLOR_BGR2RGB)
  resized = cv2.resize(RGB_img, (Img_Size,Img_Size))
  plt.imshow(resized)
  plt.title(Y_test[i])
  plt.axis("off")
```
