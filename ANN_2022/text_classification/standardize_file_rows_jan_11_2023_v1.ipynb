```python
file_name = 'ItensVitrine3P_Farmers.xlsx'
df = pd.read_excel(file_name)
df = df[['Região','NegocioCliente','CodigoProdutoSeller']]
df
```

```python
column2split = df['CodigoProdutoSeller'].values
for i in column2split:
  if(i.find(',') != -1):
    ls = i.split(",")
    for j in ls:
      print(j.replace(" ", ""))
```

```python
column2split = df['CodigoProdutoSeller'].values
regiao=df['Região'].values
NegocioCliente= df['NegocioCliente'].values
k = 0
c1, c2, c3 = [], [],[]
for i in column2split:
  if(i.find(',') != -1):
    ls = i.split(",")
    for j in ls:
      c3.append(j.replace(" ", ""))
      c1.append(regiao[k])
      c2.append(NegocioCliente[k])
  else:
    c3.append(i.replace(" ", ""))
    c1.append(regiao[k])
    c2.append(NegocioCliente[k])
  k += 1
df_final = pd.DataFrame({'Região':c1,'NegocioCliente':c2,'CodigoProdutoSeller':c3  })
```
