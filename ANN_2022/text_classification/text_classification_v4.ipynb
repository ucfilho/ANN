```python
def CountTextDFsample(df_new,df_classify,sample):
  # sample: max number to select
  text2category= df_new['text2category'].values
  category_labels = df_new['category_labels'].values
  w = df_classify['text info'].values
  
  common_list =[]
  idx = 0
  for j in w:
    list_01 =[]
    k, name  = 0, 'none'
    for i in text2category:
      common_number = len(set(i).intersection(j))
      name=category_labels[k]
      k += 1
      list_01.append([common_number,name ])
    
    idx, list_02 = 0, []
    list_01.sort(reverse=True)
    print(list_01)
    for k in list_01:
      if(idx < sample):
        list_02.append(k)
      idx += 1

    common_list.append(list_02)
    
    
  return common_list
```
