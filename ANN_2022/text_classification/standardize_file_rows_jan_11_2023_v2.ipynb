```python
file_name = 'ItensVitrine3P_Farmers.xlsx'
df = pd.read_excel(file_name)
df = df.iloc[:,[0,1,2]]
df
```

```python
def convert_default_format(df):
  names = df.columns
  column2split = df[names[2]].values
  regiao=df[names[0]].values
  NegocioCliente= df[names[1]].values
  k = 0
  c1, c2, c3 = [], [],[]
  for i in column2split:
    if(i.find(',') != -1):
      ls = i.split(",")
      for j in ls:
        c3.append(j.replace(" ", ""))
        c1.append(regiao[k])
        c2.append(NegocioCliente[k])
    else:
      c3.append(i.replace(" ", ""))
      c1.append(regiao[k])
      c2.append(NegocioCliente[k])
    k += 1
  df_final = pd.DataFrame({'Região':c1,'NegocioCliente':c2,'CodigoProdutoSeller':c3  })
  return df_final
```

```python
df_final=convert_default_format(df)
df_final.head(12)
```
