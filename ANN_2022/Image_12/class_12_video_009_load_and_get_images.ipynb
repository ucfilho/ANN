```python
def find2lines(get_edges):
  y = []
  for vi_edges in get_edges:
    cont = 0
    for vi in vi_edges:
      if vi == True:
        cont += 1
    y.append(cont)
  return y
```

```python
name_video = 'strawberry.mp4'
cap = cv2.VideoCapture(name_video)
cont = 0 
filtro =  100
if cap.isOpened() == False:
  print('video not found')

k = 0
images=[]

while cap.isOpened():
  cont = cont + 1
  ret,frame =cap.read()

  if ret == True:
    if(cont % filtro == 0):
      
      images.append(frame)
      k += 1
      gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
      edges = feature.canny(gray, sigma=1)
      y = find2lines(edges)
      n = len(y)

      for i in range(n):
        if( y[i] < 10 and y[i]>5):
          cv2.line(frame, (100,i), (400,i), (255,0,0), 1)
      
      cv2_imshow(frame)


      
    if cv2.waitKey(10) & 0xFF == ord('q'):
      break
  else:
    break


cap.release()
cv2.destroyAllWindows()
```
