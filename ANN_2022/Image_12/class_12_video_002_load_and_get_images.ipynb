```python
name_video = 'strawberry.mp4'
cap = cv2.VideoCapture(name_video)
cont = 0 
filtro =  500
if cap.isOpened() == False:
  print('video not found')

k = 0
images=[]
img_names = []

while cap.isOpened():
  cont = cont + 1
  ret,frame =cap.read()

  if ret == True:
    if(cont % filtro == 0):
      cv2_imshow(frame)
      images.append(frame)
      if(k<10):
        name = path+'/' + 'frame_0'+str(k)+'.jpg'
      else:
        name = path + '/' + 'frame_'+str(k)+'.jpg'
      img_names.append(name)
      k += 1
      result=cv2.imwrite(name, frame)

    if cv2.waitKey(10) & 0xFF == ord('q'):
      break
  else:
    break


cap.release()
cv2.destroyAllWindows()
```
