```python
import cv2
from google.colab.patches import cv2_imshow
from skimage import feature
import numpy as np
```

```python
name_video = 'strawberry.mp4'
cap = cv2.VideoCapture(name_video)
cont = 0 
filtro =  100
if cap.isOpened() == False:
  print('video not found')

k = 0
images=[]

while cap.isOpened():
  cont = cont + 1
  ret,frame =cap.read()

  if ret == True:
    if(cont % filtro == 0):
      
      images.append(frame)
      k += 1
      gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
      edges = feature.canny(gray, sigma=1)
      nrows,ncols =edges.shape
      Foto = np.ones((nrows,ncols))*255
      for i in range(nrows):
        for j in range(ncols):
          if edges[i,j] == True:
            Foto[i,j]=0
      cv2_imshow(Foto)

      if(k==5):
        get_edges = edges
        pic = frame
      
    if cv2.waitKey(10) & 0xFF == ord('q'):
      break
  else:
    break


cap.release()
cv2.destroyAllWindows()
```

```python
def find2lines(get_edges):
  y = []
  for vi_edges in get_edges:
    cont = 0
    for vi in vi_edges:
      if vi == True:
        cont += 1
    y.append(cont)
  return y
```

```python
nrows,ncols =get_edges.shape
Foto = np.ones((nrows,ncols))*255
y = find2lines(get_edges)
n = len(y)

for i in range(n):
  if( y[i] < 5 and y[i]>2):
    Foto[i,range(ncols)]=0

cv2_imshow(Foto)
cv2_imshow(pic)
```
