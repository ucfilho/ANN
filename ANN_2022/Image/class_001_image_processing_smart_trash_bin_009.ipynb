```python
f1='john_mccarthy.png'
Img_Size = 400
img1 = cv2.cvtColor(cv2.imread(f1), cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(10, 10))

#img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
resized = cv2.resize(img1, (Img_Size,Img_Size))
plt.imshow(resized)
plt.title(f1)
plt.axis("off")
```

```python
img1.shape
```

```python
width = 300
img=np.zeros((width,width,3)).astype(np.int64)
for i in range(width):
  for j in range(width):
    for k in range(3):
       img[i,j,k] = img1[i,j,k]
```

```python
plt.figure(figsize=(10, 10))


plt.imshow(img)
plt.title(f1)
plt.axis("off")
```
