```python
f1='john_mccarthy.png'
Img_Size = 400
img1 = cv2.imread(f1, 0)
```

```python
plt.figure(figsize=(10, 10))

#img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
resized = cv2.resize(img1, (Img_Size,Img_Size))
plt.imshow(resized, cmap='gray')
plt.title(f1)
plt.axis("off")
```

```python
# https://scikit-image.org/docs/stable/api/skimage.feature.html#skimage.feature.greycoprops
```

```python
g = skimage.feature.greycomatrix(img1, [1, 2], [0, np.pi/2], levels=255,normed=True, symmetric=True)
contrast = skimage.feature.greycoprops(g, 'contrast')
print(contrast)
```
