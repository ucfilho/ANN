```python
FILE='trash_nov_22_2018.zip'
```

```python
archive = zipfile.ZipFile('trash_nov_22_2018.zip', 'r')
archive.extractall()
```

```python
img_name=[]
with zipfile.ZipFile(FILE, "r") as f:
    for name in f.namelist():
        img_name.append(name)
```

```python
Img_Size = 80
ref = 'met'
name = 'xwz'
```

```python
X_test= []  
Y_test =[]
n = len(img_name)
i =0

while(i<16):
  name = img_name[random.randrange(n)]
  if(ref in name):
    i = i + 1
    Y_test.append(name)
    X_test.append(cv2.cvtColor(cv2.imread(name), cv2.COLOR_BGR2RGB))
```

```python
plt.figure(figsize=(10, 10))

for i in range(16): 
  ax = plt.subplot(4, 4, i + 1)
  resized = cv2.resize(X_test[i], (Img_Size,Img_Size))
  plt.imshow(resized)
  plt.title(Y_test[i])
  plt.axis("off")
```
