```python
f1='john_mccarthy.png'
Img_Size = 400
img1 = cv2.cvtColor(cv2.imread(f1), cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(10, 10))

#img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
resized = cv2.resize(img1, (Img_Size,Img_Size))
plt.imshow(resized)
plt.title(f1)
plt.axis("off")
```

```python
gauss = cv2.GaussianBlur(img1, (5,5), 1,1)
```

```python
plt.figure(figsize=(10, 10))

#img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
resized = cv2.resize(gauss, (Img_Size,Img_Size))
plt.imshow(resized)
plt.title(f1)
plt.axis("off")
```
