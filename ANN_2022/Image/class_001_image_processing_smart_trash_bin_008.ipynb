```python
f1='arvore.jpg'
f2='john_mccarthy.png'
Img_Size = 400
img1 = cv2.cvtColor(cv2.imread(f1), cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(cv2.imread(f2), cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(10, 10))

#img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
resized = cv2.resize(img1, (Img_Size,Img_Size))
plt.imshow(resized)
plt.title(f1)
plt.axis("off")
```

```python
plt.figure(figsize=(10, 10))

RGB_img = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
resized = cv2.resize(RGB_img, (Img_Size,Img_Size))
plt.imshow(resized)
plt.title(f2)
plt.axis("off")
```
