```python
import cv2
import skimage
import skimage.feature
from scipy.signal import convolve2d
import random
import zipfile
import numpy as np
from matplotlib import pyplot as plt
```

```python
f1='john_mccarthy.png'
Img_Size = 400
img1 = cv2.imread(f1, 0)
```

```python
plt.figure(figsize=(10, 10))

#img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
resized = cv2.resize(img1, (Img_Size,Img_Size))
plt.imshow(resized, cmap='gray')
plt.title(f1)
plt.axis("off")
```

```python
# https://medium.com/swlh/image-processing-with-python-convolutional-filters-and-kernels-b9884d91a8fd
```

```python
# sobel
kernel = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])
```

```python
conv_img1 = convolve2d(img1,kernel[::-1, ::-1]).clip(0,1)
```

```python
plt.figure(figsize=(10, 10))

#resized = cv2.resize(conv_img1, (Img_Size,Img_Size))
plt.imshow(conv_img1, cmap='gray')
plt.title(f1)
plt.axis("off")
```
